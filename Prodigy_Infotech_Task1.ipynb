```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (you already did this earlier)
df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_394806.csv", skiprows=4)

# Select relevant columns: Country Name and 2022 data
df_2022 = df[["Country Name", "2022"]]

# Drop rows with missing values
df_2022 = df_2022.dropna()

# Sort and select top 10 most populated countries
top10 = df_2022.sort_values(by="2022", ascending=False).head(10)
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(top10["Country Name"], top10["2022"], color='skyblue')
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries in 2022")
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
#ddd
plt.figure(figsize=(12, 6))
plt.bar(top10["Country Name"], top10["2022"], color='orange')
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries in 2022")
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.savefig("population_bar_chart.png")  # Saves the image
plt.show()
```
